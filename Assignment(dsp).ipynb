```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-25,26)
n
```

```python
def x(n):
    if 0 <= n <= 10:
        return (n * n) + n
    elif -10 <= n < 0:
        return -(n * n) + n
    else:
        return 0

xn = [x(i) for i in n]
xn
```

```python
xn = np.array(xn)
xn
```

```python
plt.stem(n,xn)
plt.show()
```

```python
#delay
nd = n[0:n.size-5]
nd
```

```python
nd = np.array(nd + 5)
nd
```

```python
xnd = xn[0:xn.size-5]
xnd
```

```python
plt.stem(nd,xnd)
```

```python
xnd = np.array(xnd * 2)
xnd
```

```python
plt.stem(nd,xnd)
```

```python
#advance
na = n[5:n.size]
na
```

```python
na = np.array(na - 5)
na
```

```python
xna = xn[5:xn.size]
xna
```

```python
plt.stem(na,xna)
```

```python
xna = np.array(xna * 2)
xna
```

```python
plt.stem(na,xna)
```

```python
mxn = []
for i in n:
    if (i+25-5>=0 and i+25+5<51):
        mxn.append(xnd[i]*xna[i-5])

mxn
```

```python
mxn = np.array(mxn)
mxn
```

```python
nn = np.arange(-20,21)
nn
```

```python
plt.stem(nn,mxn)
```

```python
#folding
mxn = np.array(mxn*-1)
mxn
```

```python
plt.stem(nn,mxn)
```
