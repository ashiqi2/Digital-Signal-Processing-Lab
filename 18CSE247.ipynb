```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-20,21)
n
```

```python
#Function x(n)

def x(n):
    if 0 <= n <= 10:
        return (n * n) + n
    elif -10 <= n < 0:
        return -(n * n) + n
    else:
        return 0

xn = [x(i) for i in n]
xn
```

```python
xn = np.array(xn)
xn
```

```python
plt.stem(xn)
```

```python
# Delay by 5

n = n[0:n.size-5]
n
```

```python
n = np.array(n + 5)
n
```

```python
xnd = xn[0:xn.size-5]
xnd
```

```python
plt.stem(xnd)
```

```python
xnd = np.array(xnd * 2)
xnd
```

```python
plt.stem(xnd)
```

```python
#Advance by 5

n = n[5:n.size]
n
```

```python
n = np.array(n - 5)
n
```

```python
xna = xn[5:xn.size]
xna
```

```python
plt.stem(xna)
```

```python
xna = np.array(xna * 2)
xna
```

```python
plt.stem(xna)
```

```python
#Function y(n)

yn = np.array(xnd * xna)
yn
```

```python
plt.stem(yn)
```

```python
#Folding 

n = np.array(n*-1)
n
```

```python
plt.stem(yn)
plt.title('y(n) = FD[(2 * x(n + 5)) * (2 * x(n - 5))]')
plt.xlabel('---------->n')
plt.ylabel('---------->y(n)')
```
